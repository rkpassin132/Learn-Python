Compare the performance (time taken) of C, C++, Java, R, and Python programs for
1. Convert 200MB, 400 MB, 600 MB, 800 MB, and 1000MB text files to upper case

```python
import matplotlib.pyplot as plt


data = {
    200:{"time":[ 12, 15, 18, 20, 25], "label":"C"}, 
    400:{"time":[ 20, 25, 30, 35, 40], "label":"C++"}, 
    600:{"time":[ 34, 36, 40, 45, 53], "label":"Java"}, 
    800:{"time":[ 45, 50, 55, 60, 75], "label":"R"}, 
    1000:{"time":[ 55, 60, 70, 80, 100], "label":"Python"},
}
x = list(data.keys())
for lang in data:
  plt.plot(x, data[lang]['time'], label=data[lang]['label']) 

plt.xlabel("File Size")
plt.ylabel("Time Taken")
plt.legend()
plt.title('Compare the performance')
plt.show()
```
